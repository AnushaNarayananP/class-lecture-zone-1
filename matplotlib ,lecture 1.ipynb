```python
import matplotlib.pyplot as plt
```

```python
x=[1,18,8,1,18,8]
y=[12,50,250,12,50,250]
plt.plot(x,y)
plt.show()
```

```python
y=[12,50,2,18,80,25]
plt.plot(y)
plt.show()
```

```python
y=[12,50,2,18,80,25]
plt.plot(y,marker='*',ms=20,mec='r',mfc='b')
plt.show()
```

```python
y=[12,50,2,18,80,25]
plt.plot(y, 'o:r')
plt.show()
```

```python

y=[12,50,2,18,80,25]
plt.plot([y, linestyle='-.'])
plt.show()
```

```python
y=[12,50,2,18,80,25]
plt.xlabel('number of hours')
plt.ylabel('car speed')
plt.plot(y)
plt.show()
```

```python
y=[12,50,2,18,80,25]
plt.xlabel('number of hours',loc='right')
plt.ylabel('car speed',loc='top')
plt.plot(y)
plt.show()
```

```python
y=[12,50,2,18,80,25]
plt.xlabel('number of hours',loc='right')
plt.ylabel('car speed',loc='top')
plt.title('Car Speed Prediction')
plt.plot(y)
plt.show()
```

```python
y=[12,50,2,18,80,25]
plt.xlabel('number of hours',loc='right')
plt.ylabel('car speed',loc='top')
plt.title('Car Speed Prediction')
plt.grid()
plt.plot(y)
plt.show()
```

```python
y=[12,50,2,18,80,25]
plt.xlabel('number of hours',loc='right')
plt.ylabel('car speed',loc='top')
plt.title('Car Speed Prediction')
plt.grid(axis='y',linewidth=2)
plt.plot(y)
plt.show()
```

```python
y=[12,50,2,18,80,25]
plt.subplot(2,2,1)
plt.plot(y)
plt.xlabel('number of hours',loc='right')
plt.ylabel('car speed',loc='top')
plt.title('Car Speed Prediction')
plt.grid()

y1=[12,50,2,18,80,25]
plt.subplot(2,2,2)
plt.plot(y1)
plt.xlabel('number of hours',loc='right')
plt.ylabel('car speed',loc='top')
plt.title('Car Speed Prediction')
plt.grid()

y2=[12,50,2,18,80,25]
plt.subplot(2,2,3)
plt.plot(y2)
plt.xlabel('number of hours',loc='right')
plt.ylabel('car speed',loc='top')
plt.title('Car Speed Prediction')
plt.grid()


y3=[12,50,2,18,80,25]
plt.subplot(2,2,4)
plt.plot(y3)
plt.xlabel('number of hours',loc='right')
plt.ylabel('car speed',loc='top')
plt.title('Car Speed Prediction')
plt.grid()
plt.suptitle(' 4 info details')
plt.show()


```
