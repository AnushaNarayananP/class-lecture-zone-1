```python
x=[55,15,10,20]
```

```python
plt.pie(x)
plt.show()
```

```python
plt.pie(x,labels=['honda','volvo','lexus','bmw'])
plt.show()
```

```python
mylabels=['honda','volvo','lexus','bmw']
```

```python
plt.pie(x,labels=mylabels)
plt.show()
```

```python
def pie(a,b):
    plt.pie(a,labels=b)
    plt.show()
pie(x,mylabels)    
```

```python
def pie(a,b):
    try:
        plt.pie(a,labels=b)
        plt.show()
    except Exception as e:
        print(e)
pie(x,mylabels)        
```

```python
def pie(a,b):
    try:
        plt.pie(a,labels=b,startangle=90)
        plt.show()
    except Exception as e:
        print(e)
pie(x,mylabels)    
```
